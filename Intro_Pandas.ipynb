```python
import urllib.request
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/OpenAPC/openapc-de/master/data/fuberlin/APC_FU_Berlin_2015.csv"
apc_file = "openapc.csv"
```

```python
# Download der Daten und lokales Speichern
urllib.request.urlretrieve(url, apc_file)
```

```python
# Daten in Pandas einlesen
apcs = pd.read_csv(apc_file)
```

```python
# Graphik abspeichern (für Weitergabe) mit Code: Variable vergeben
hist_euro = apcs["euro"].hist(bins=20, color="turquoise")
```

```python
# Graphik abspeichern (für Weitergabe) mit Code: get figure ("in die Hand drücken")
figure = hist_euro.get_figure()
```

```python
# Graphik abspeichern (für Weitergabe) mit Code: speichern
figure.savefig("apc_hist.png")
```
